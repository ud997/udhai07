```python
#importing all the files from the given two excel files
File=pd.read_excel("data.xlsx",sheet_name=None)
File1=pd.read_excel("data_1.xlsx",sheet_name=None)
```

```python
#itrating through all the sheets in the excel data.xlsx
list1=[]
dataframes=[]
dataframes1=[]
dataframes2=[]
keys = File.keys()
for i in keys:
  list1.append(i)
for i in list1:
    if 'Detail_67' in i:
        dataframes.append(File[i])
    if 'DetailVol_67_' in i:
        dataframes1.append(File[i])
    if 'DetailTemp_67_' in i:
        dataframes2.append(File[i])
```

```python
# itrating throufh the second excel file dara_1.xlsx
list2=[]
keys1= File1.keys()
for i in keys1:
  list2.append(i)
for i in list2:
    if 'Detail_67' in i:
        dataframes.append(File1[i])
    if 'DetailVol_67_' in i:
        dataframes1.append(File1[i])
    if 'DetailTemp_67_' in i:
        dataframes2.append(File1[i])
join=pd.concat(dataframes)
join.to_csv('detail.csv')
join1=pd.concat(dataframes1)
join1.to_csv('detailVol.csv')
join1=pd.concat(dataframes2)
join1.to_csv('detailTemp.csv')
```
